```python
from os.path import dirname, join
import numpy as np


from bokeh.plotting import figure
from bokeh.layouts import layout, widgetbox
from bokeh.models import ColumnDataSource, Div
from bokeh.models.widgets import Slider, Select, TextInput
from bokeh.io import curdoc

columns = stations.tolist()

path = '/home/juhak/projections/hls_bikez/'
filename = "bike_meter.html"

desc = Div(text=open(join(path, filename)).read(), width=800)

axis_map = {
    "empty_slots": "empty_slots",
    "free_bikes": "free_bikes",
    "total_slots": "total_slots"
}

# Create Input controls
station_name = Select(title="Select station", value="Varsapuistikko", options=columns)
y_axis = Select(title="Y Axis", options=sorted(axis_map.keys()), value="empty_slots")

# Create Column Data Source that will be used by the plot
source = ColumnDataSource(data=dict(x=[], y=[]))

p = figure(plot_height=600, plot_width=700, title="", toolbar_location=None)
p.circle(x="x", y="y", source=source, size=7, line_color=None)


def select_station():
    station_val = station_name.value
    selected = df[(df.name == station_val)]
    return selected, station_val

def update():
    sel, station = select_station()
    x_name = 'timestamp'
    y_name = axis_map[y_axis.value]
    
    p.yaxis.axis_label = y_axis.value
    print(sel)
    source.data = dict(
        x=sel[x_name],
        y=sel[y_name]
    )
    
controls = [station_name]
for control in controls:
    control.on_change('value', lambda attr, old, new: update())
    
sizing_mode = 'fixed' # 'scale_width' also looks nice with this example

inputs = widgetbox(*controls, sizing_mode=sizing_mode)
l = layout([
    [desc],
    [inputs, p],
], sizing_mode=sizing_mode)

update()

curdoc().add_root(l)
curdoc().title = "Movies"
```
